```python
import pandas as pd
import plotly.express as px
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("gilbert-regression.csv", header=None, names=["Filename", "Category", "X", "Y"])

# Create a scatter plot
fig = px.scatter(
    df,
    x="X",
    y="Y",
    color="Category",  # Color points by category
    opacity=0.2,      # Set transparency
    log_x=True,       # Set x-axis to log scale
    title="",
    labels={"X": "P(Received to Lost)", "Y": "P(Lost to Received)"}
)
fig.update_layout(
    autosize=False,
    width=600,  # Set width of the plot
    height=400,  # Set height of the plot
    margin=dict(l=40, r=40, t=40, b=40)  # Set margins
)

# Show the plot
fig.show()
```
